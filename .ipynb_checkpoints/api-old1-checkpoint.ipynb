```python
def download_pop(state_name):
    'auto download population by statename'
    url = 'https://fred.stlouisfed.org/series/'+ state_name+'POP'
    print(url)
    
    # open the website
    browser.get(url)
    time.sleep(10)
    
    button = browser.find_element_by_id('download-button')
    button.click()
    
    time.sleep(10)
    download = browser.find_element_by_id('download-data-csv')
    download.click()
```

```python
len(state_names)
```

# Temperature API

From https://www.ncei.noaa.gov/support/access-data-service-api-user-documentation 

Interested in TAVG, TMAX, TMin and PRCP for each station (daily) 


Example 
https://www.ncei.noaa.gov/access/services/data/v1?dataset=daily-summaries&stations=USC00457180&startDate=1990-01-01&endDate=2019-02-28&dataTypes=TAVG,TMAX,TMIN,PRCP&format=csv&includeAttributes=False&includeStationName=true&includeAttributes=0&includeStationName=true


```python
# Load station ID files
stations = pd.read_csv('E:/Users/Fern/energy_data/US_temp/ushcn_stations.csv',header=None)
stations.head()
```

```python
station_ids = stations.iloc[:,0].values
station_ids
```

```python
browser = webdriver.Firefox() 
```

```python
def download_temp(name):
    'auto download temperature by station name'
    url = f'https://www.ncei.noaa.gov/access/services/data/v1?dataset=daily-summaries&stations={name}&startDate=1990-01-01&endDate=2019-02-28&dataTypes=TAVG,TMAX,TMIN,PRCP&format=csv&includeAttributes=False&includeStationName=true&includeAttributes=0&includeStationName=true'
    #print(url)
    
    # open the website
    browser.get(url)
    time.sleep(10)
```

```python
for station_name in station_ids:
    download_temp(station_name)
```
